```python

import matplotlib.pyplot as plt
import os

filename = os.path.join("data/distance_30cm_fist_bump", "1713059691_5s_data.txt") 

with open(filename, "r") as file:
    data = [int(line.strip()) for line in file if line.strip()]

plt.figure(figsize=(10, 6))
plt.plot(data[-7600:-7000], marker='o', linestyle='-', markersize=4, label='Sensor Data')  #可视化一个拳头敲击周期的动作
plt.title('Sensor Data Over Time')
plt.xlabel('Time (arbitrary units)')
plt.ylabel('Sensor Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python

import matplotlib.pyplot as plt
import os

filename = os.path.join("data/distance_30cm_fist_bump", "1713059697_5s_data.txt") 

with open(filename, "r") as file:
    data = []
    for line in file:
        stripped_line = line.strip()
        if stripped_line:  
            try:
                # try to convert the line to an integer
                data.append(int(stripped_line))
            except ValueError:
                # handle the case where the line contains non-integer values
                parts = stripped_line.split('\x00')
                for part in parts:
                    if part:  
                        try:
                            data.append(int(part))
                        except ValueError:
                            print(f"warning: '{part}' cannot be converted to an integer")

plt.figure(figsize=(10, 6))
plt.plot(data[1500:3000], marker='o', linestyle='-', markersize=4, label='Sensor Data') #可视化一个拳头敲击周期的动作
plt.title('Sensor Data Over Time')
plt.xlabel('Time (arbitrary units)')
plt.ylabel('Sensor Value')
plt.legend()
plt.grid(True)
plt.show()
```
