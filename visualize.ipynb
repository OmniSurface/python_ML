```python

import matplotlib.pyplot as plt
import os

# 假设数据文件位于"data"文件夹下，文件名为示例中提到的格式。
# 这里我们只读取一个文件进行演示。如果有多个文件，可以用循环遍历"data"目录下的所有文件。
filename = os.path.join("data", "1713059691_5s_data.txt")  # 替换your_data_file.txt为实际的文件名

# 读取数据
with open(filename, "r") as file:
    data = [int(line.strip()) for line in file if line.strip()]

# 数据可视化
plt.figure(figsize=(10, 6))
plt.plot(data, marker='o', linestyle='-', markersize=4, label='Sensor Data')
plt.title('Sensor Data Over Time')
plt.xlabel('Time (arbitrary units)')
plt.ylabel('Sensor Value')
plt.legend()
plt.grid(True)
plt.show()
```
